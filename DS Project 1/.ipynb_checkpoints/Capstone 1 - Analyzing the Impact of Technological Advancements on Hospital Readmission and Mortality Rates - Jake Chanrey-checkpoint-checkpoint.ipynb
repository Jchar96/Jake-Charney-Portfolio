```python
# importing all neseccary Pyhton packages

import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
# laading data into pannds dataframe 

hospital_data = pd.read_csv('Readmissions_and_Deaths.csv')
```

```python
hospital_data.head()
```

```python
hospital_data = hospital_data.set_index('index')
hospital_data.head()
```

```python
hospital_data.shape
```

```python
hospital_data.info()
```

```python
hospital_data[hospital_data['Denominator'] == 'Not Available']
```

```python
# dropping rows with missing numeric data

hospital_data = hospital_data[hospital_data['Denominator'] != 'Not Available'] 
```

```python
hospital_data['Denominator'].unique()
```

```python
hospital_data.dtypes
```

```python
# converting numeric data to float64 dtype 

hospital_data['Denominator'] = pd.to_numeric(hospital_data['Score'])
hospital_data['Score'] = pd.to_numeric(hospital_data['Score'])
hospital_data['Lower Estimate'] = pd.to_numeric(hospital_data['Lower Estimate'])
hospital_data['Higher Estimate'] = pd.to_numeric(hospital_data['Higher Estimate'])
```

```python
print(hospital_data.shape)
print(hospital_data.dtypes)
```

```python
# changing data type of start data and end date to datetime dtype

hospital_data['Measure Start Date'] = pd.to_datetime(hospital_data['Measure Start Date'])
hospital_data['Measure End Date'] = pd.to_datetime(hospital_data['Measure End Date'])
```

```python
hospital_data.dtypes
```

```python
# generateding a data frame to determine if there are any more missing values. 

missing = pd.concat([hospital_data.isnull().sum(), 100 * hospital_data.isnull().mean()], axis=1)
missing.columns=['count', '%']
missing.sort_values(['count', '%'])
```

```python
missing['count'].plot(kind='hist', bins=20, title='count')
plt.gca().spines[['top', 'right',]].set_visible(False)
plt.xlabel('missing value count')
```

```python
# removing columns irrelevant to data analysis

hospital_data = hospital_data.drop(columns=['Phone Number','Footnote'], axis=1)
```

```python
print(hospital_data.columns)
```

```python
hospital_data['Measurement Period Length'] = (hospital_data['Measure End Date'] - hospital_data['Measure Start Date']).dt.days
```

```python
print(hospital_data['Measurement Period Length'].head(10))
print(hospital_data['Measurement Period Length'].unique().sum())
```
